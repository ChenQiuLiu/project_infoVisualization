```python
def getfirst(ls):
    return ls[0]
def scatterGTwoVar(data, normx = False, normy = True):
    '''
    draw scatter about two factor to show relation.
    If the values of independent variables are the same, the maximum and minimum values are marked.
    
    @parameter
        data: list, tuple. data[0] is  dependent var. data[1] is  independent var.
              data must 
        norm: x,y need normlization ?
    
    '''
    dic = {key:[] for key in set(list(data[0]))}
    line_max = []
    line_min = []
    for i in range(len(list(data[0]))):
        dic[data[0].iloc[i]].append(data[1].iloc[i])
    for i in dic.keys():
        for j in dic[i]:
            if j == max(dic[i]):
                plt.plot(i, j, 'o',c = 'red', alpha=0.4)
                line_max.append((i, j))
            elif j == min(dic[i]):
                plt.plot(i, j, 'o', c = 'black', alpha=0.4)
                line_min.append((i, j))
            else:
                plt.plot(i, j, 'o', c = 'blue', alpha=0.4)
    line_max.sort(key = getfirst)
    line_min.sort(key = getfirst)
    plt.plot([i[0] for i in line_max], [i[1] for i in line_max], c = 'red', label = 'max in each 5 items')
    plt.plot([i[0] for i in line_min], [i[1] for i in line_min], c = 'green', label = 'min in each 5 items') 
    plt.fill_between([i[0] for i in line_max], [i[1] for i in line_max], [i[1] for i in line_min], alpha = 0.2)
    plt.legend()    
    plt.show()

a = df21["no_features"][data["RQ"] == 2.2]
b = df21["train_energy(J)"][data["RQ"] == 2.2]
scatterGTwoVar([a, b])
```
